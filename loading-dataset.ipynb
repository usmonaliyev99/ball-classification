```python
import pandas as pd
from pathlib import Path
from glob import glob

CURRENT_DIR = Path().resolve()

images = glob(f"{CURRENT_DIR}/OIDv4_ToolKit/OID/Dataset/train/*/*.jpg")

df = pd.DataFrame({'path': images})

df.sample(10)
```

```python
labels = ['Football', 'Tennis ball', 'Volleyball']

def put_label(df):
    for label in labels:
        df.loc[df.path.str.contains(label, False), 'label'] = label

    return df
```

```python
df = put_label(df)
```

```python
def filename(name: str):
    
    return name[name.rindex('/')+1:]
    
df['name'] = df.path.map(filename)

df
```

```python
import os

def loc(path:str):
    index = path.rindex('/') + 1
    name = path[index:-4]
    core = path[:index]

    loc = f"{core}Label/{name}.txt"
    if os.path.exists(loc):
        with open(loc) as f:
            return f.read()

df['loc'] = df.path.apply(loc)

df
```

```python
df['loc'] = df['loc'].map(lambda x: x[x.index('ball ')+5:])
```

```python
df['loc'] = df['loc'].map(lambda x: x[:x.index('\n')])
```

```python
df['loc'] = df['loc'].map(lambda x: x.split(' '))
```

```python
def cut_image(image, loc):
    loc = [int(float(x)) for x in loc]
    x1, y1, x2, y2 = loc

    cX, cY = x1 + int((x2 - x1) / 2), y1 + int((y2 - y1) / 2)
    x1, y1, x2, y2 = cX - 120, cY - 120, cX + 120, cY + 120

    x1 = x1 if x1 > 0 else 0
    x2 = x2 if x2 > 0 else 0
    y2 = y2 if y2 > 0 else 0
    y1 = y1 if y1 > 0 else 0

    return image[y1:y2, x1:x2]
```

```python
import os
import matplotlib.pyplot as plt


for index, item in df.iterrows():
    img = plt.imread(item.path)    
    img = cut_image(img, item['loc'])

    if not os.path.exists(f"./data/{item.label}"):
        os.mkdir(f"./data/{item.label}")
    
    plt.imsave(f"./data/{item.label}/{item['name']}", img)

```
